```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = "/content/sample_data/Iris.csv"
df = pd.read_csv(path)
df.tail(5)
```

```python
df["Species"].value_counts()
```

```python
sns.FacetGrid(df,hue="Species", height = 6).map(plt.scatter, "SepalLengthCm","Species").add_legend()
```

```python
flower_mapping = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}
df["Species"] = df["Species"].map(flower_mapping)
```

```python
df.tail()
```

```python
X = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
y = df[['Species']].values
y= y.ravel()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
expected = y
predicted = model.predict(X)
print("Pridicted: ")
predicted
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(expected,predicted))
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
```

```python
svmMod = SVC(kernel='linear')
svmMod.fit(X,y)
```

```python
expct = y
predict = svmMod.predict(X)
```

```python
print(accuracy_score(expct, predict))
print(classification_report(expct, predict))
```
